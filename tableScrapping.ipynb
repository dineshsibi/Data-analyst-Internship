```python
import requests
```

```python
from bs4 import  BeautifulSoup
```

```python
url='https://www.forbesindia.com/article/explainers/top-10-it-companies-in-india/87143/1'
```

```python
it=requests.get(url)
```

```python
soup=BeautifulSoup(it.text,'html')
```

```python
print(soup)
```

```python
table=soup.find('table')
```

```python
table
```

```python
heading=table.find_all('th')
```

```python
heading
```

```python
heading_text=[title.text for title in heading]
```

```python
heading_text
```

```python
pip install pandas
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(columns=heading_text)
```

```python
df
```

```python
table_cont=table.find_all('tr')
```

```python
for topIt in table_cont[1:]:
  row_content=topIt.find_all('td')
  row_data_cont=[data.text.strip() for data in row_content]

  length=len(df)
  df.loc[length]=row_data_cont
```
